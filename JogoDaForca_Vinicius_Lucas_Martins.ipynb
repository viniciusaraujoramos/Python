```python
#Lucas Martins Pereira e Vinicius Araujo Ramos

from getpass import getpass

palavraSecreta = ""
letrasPalavra = []
espacos = []
vidas = 5
status = ""


#Parte do código onde é feita a introdução do jogo e a extração da palavra secreta
def programa():
  apresentacao()

def apresentacao():
  global palavraSecreta
  print("Bem vindo ao jogo da forca do Lulu e Vcin")
  nome = input("Digite o seu nome:")
  palavraSecreta = getpass("Digite a palavra secreta:").upper()
  if len(nome) <= 2:
   print("Digite o seu nome corretamente, ou por extenso")
   apresentacao()
  else:
    preparacao(nome)

#Preenche a lista com _ de acordo com a quantidade de letras da palavra
def preparacao(nome):
  global espacos
  global letrasPalavra
  for letras in palavraSecreta:
    espacos.append("_")
    letrasPalavra.append(letras)
  inicio(nome)

#Mensagem de boas vindas ao jogador e mostra a quantidade de espaços vazios
def inicio(nome):
  global vidas
  global espacos
  print("Bem vindo",nome + "! Vamos começar")
  print("Você possui ",vidas," vidas!")
  for espaco in espacos:
    print(espaco," ",end="")
  letraDigitada()

#Mensagem que solicita o jogador a digitar uma letra, deixa ela maiúscula
def letraDigitada():
  letraAtual = input("Digite uma letra:").upper()
  if len(letraAtual)!= 1:
    print("Por favor, digite apenas uma letra")
    letraDigitada()
  else:
    verificacao(letraAtual)

#Verifica se a letra atual digitada pelo jogador está na palavra secreta.
#Atualiza os espacos com a letra correta e verifica se o jogador ganhou ou perdeu.
def verificacao(letraAtual):
  global vidas
  global espacos
  contador = 0
  status = "erro"
  for letra in letrasPalavra:
    if letraAtual == letra:
      status = "acerto"
      espacos[contador] = letraAtual
    contador = contador + 1

  if status == "acerto":
    print("Existe essa letra")
    if espacos == letrasPalavra:
      print("Você ganhou")
  else:
    print("Não existe essa letra!")
    vidas -=1
    if vidas == 0:
      print("Você perdeu")
      print("A palavra era",palavraSecreta)
      exit()
    else:
      print("Você ainda tem ", vidas," vidas(s)!")


  for espaco in espacos:
    print(espaco, " ", end="")
  letraDigitada()

programa()
```
